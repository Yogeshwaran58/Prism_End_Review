```python
import os
import glob

dir_texts = r"F:\prism\Final_Prism_Dataset\val_text - Copy"   
dir_images = r"F:\prism\Final_Prism_Dataset\val_image - Copy"  

text_files = {os.path.splitext(os.path.basename(f))[0] for f in glob.glob(os.path.join(dir_texts, '*.txt'))}
image_files = {os.path.splitext(os.path.basename(f))[0] for f in glob.glob(os.path.join(dir_images, '*.*'))}  

texts_without_images = text_files - image_files

for filename in texts_without_images:
    filepath = os.path.join(dir_texts, filename + '.txt')
    os.remove(filepath)
    print(f"Deleted {filepath}")

print("Finished processing.")
```
