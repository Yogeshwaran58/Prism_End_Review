```python
import os

dir1 = r"F:\prism\Final_Prism_Dataset\validation_json - Copy"
dir2 = r"F:\prism\Final_Prism_Dataset\json - Copy"

def delete_common_files(dir1, dir2):
    
    files_in_dir1 = set(os.listdir(dir1))
    files_in_dir2 = set(os.listdir(dir2))

    
    common_files = files_in_dir1.intersection(files_in_dir2)

    
    for file in common_files:
        file_path = os.path.join(dir2, file)
        try:
            os.remove(file_path)
            print(f"Deleted {file_path}")
        except OSError as e:
            print(f"Error deleting {file_path}: {e}")

if __name__ == "__main__":
    delete_common_files(dir1, dir2)
```
