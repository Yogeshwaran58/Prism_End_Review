```python
import os
import json
import shutil
from pathlib import Path

images_dir = Path(r"F:\Icon Detection\myvenv\datasets\images\total")
json_dir = Path(r"F:\Icon Detection\myvenv\final_annotations\Total")
filtered_images_dir = Path(r"C:\content\prism\destination")
filtered_images_dir.mkdir(exist_ok=True)  

excluded_icons = {'arrow forward', 'menu', 'more', 'arrow backward'}

for json_file in json_dir.glob("*.json"):
    with open(json_file, 'r') as f:
        metadata = json.load(f)

    
    if any(icon['name'] in excluded_icons for icon in metadata.get('outputs', {}).get('object', [])):
        continue  

    
    image_name = json_file.stem + '.jpg'
    source_image_path = images_dir / image_name
    destination_image_path = filtered_images_dir / image_name

    if source_image_path.exists():
        shutil.copy(source_image_path, destination_image_path)
        print(f"Copied {image_name} to the filtered directory.")

print("Filtering complete.")
```
