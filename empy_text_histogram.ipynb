```python
import os
import glob
import matplotlib.pyplot as plt

def plot_empty_vs_nonempty_histogram(directory):
    
    if not directory.endswith('/'):
        directory += '/'

    
    text_files = glob.glob(directory + '*.txt')

    
    empty_count = 0
    non_empty_count = 0

    for file_path in text_files:
        
        if os.path.getsize(file_path) == 0:
            empty_count += 1
        else:
            non_empty_count += 1

    
    categories = ['Empty', 'Non-Empty']
    counts = [empty_count, non_empty_count]

    
    plt.figure(figsize=(8, 6))
    plt.bar(categories, counts, color=['red', 'blue'])
    plt.title('Number of Empty vs Non-Empty Text Documents_old')
    plt.xlabel('Document Type')
    plt.ylabel('Number of Documents')
    plt.xticks(categories)
    plt.show()
    print(counts)


directory = r"F:\Icon Detection\myvenv\datasets_1\labels\train"  
plot_empty_vs_nonempty_histogram(directory)
```
