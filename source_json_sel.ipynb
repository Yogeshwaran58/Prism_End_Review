```python
import os
import shutil
from pathlib import Path

images_dir = Path(r"C:\content\prism\source")
json_dir = Path(r"F:\Icon Detection\myvenv\final_annotations\Total")
filtered_json_dir = Path(r"C:\content\prism\source_json")
filtered_json_dir.mkdir(exist_ok=True)  

for image_file in images_dir.glob("*.jpg"):
    json_filename = image_file.stem + '.json'
    source_json_path = json_dir / json_filename
    destination_json_path = filtered_json_dir / json_filename
    
    if source_json_path.exists():
        shutil.copy(source_json_path, destination_json_path)
        print(f"Copied {json_filename} to the filtered JSON directory.")

print("Fetching corresponding JSON files complete.")
```
